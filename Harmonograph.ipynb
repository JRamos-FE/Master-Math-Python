Creating and displaying a harmonograpoh

```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Constants
n = 1000000
t = np.logspace(np.log10(10),np.log10(500), n)

#Parameters
A = [3, 2, 4, 4]
d = [.006, .001, .004, .0015]
f = [3, 1, 2, 2.5]

#Generating X,Y value pairs
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

#Plotting
plt.plot(x,y,'k',linewidth=.1)
plt.axis('off')
plt.show()
```

```python
#Plotting the elementary components
plt.plot(np.sin(t*f[0])*np.exp(-d[0]*3*t),linewidth=.5)
plt.show()
```
